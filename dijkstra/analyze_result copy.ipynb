```python
import pandas as pd
import seaborn as sns


from matplotlib import pyplot as plt
```

```python
folder = ""
graph = "graph_10_nodes"

df = pd.read_csv(f"rank_effect_{graph}.csv")
```

```python
df.head()
```

```python
df['CVF (Avg)'] = df['CVF In (Avg)'] + df['CVF Out (Avg)']
df['CVF (Max)'] = df['CVF In (Max)'] + df['CVF Out (Max)']
```

```python
fig, axs = plt.subplots(2, 2, figsize=(10, 6),  constrained_layout=True)

sns.lineplot(data=df, x='Rank Effect', y='PT (Avg)', ax=axs[0, 0])
axs[0, 0].set(xlabel="Rank Effect", ylabel="Count")
axs[0, 0].set_title("Program Transition Avg Effect")
axs[0, 0].set_yscale('log')

sns.lineplot(data=df, x='Rank Effect', y='PT (Max)', ax=axs[0, 1])
axs[0, 1].set(xlabel="Rank Effect", ylabel="Count")
axs[0, 1].set_title("Program Transition Max Effect")
axs[0, 1].set_yscale('log')

sns.lineplot(data=df, x='Rank Effect', y='CVF (Avg)', ax=axs[1, 0])
axs[1, 0].set(xlabel="Rank Effect", ylabel="Count")
axs[1, 0].set_title("CVF Avg Effect")
axs[1, 0].set_yscale('log')

sns.lineplot(data=df, x='Rank Effect', y='CVF (Max)', ax=axs[1, 1])
axs[1, 1].set(xlabel="Rank Effect", ylabel="Count")
axs[1, 1].set_title("CVF Max Effect")
axs[1, 1].set_yscale('log')
```

```python
pt_rank_df = pd.read_csv(f"{folder}pt_rank_{graph}.csv")

pt_rank_ravg_counts = pt_rank_df['Ar'].value_counts()
pt_rank_max_counts = pt_rank_df['M'].value_counts()
```

```python
print()
print("Program Trans Count:")
print("--------------")
print(
    "{0: >25}{1: >25}{2: >25}".format("Rank", "Avg", "Max")
)
for i in sorted(set(pt_rank_ravg_counts.index) | set(pt_rank_max_counts.index)):
    print("{0: >25}{1: >25}{2: >25}".format(i, pt_rank_ravg_counts.get(i, '0'), pt_rank_max_counts.get(i, '0')))

print()
print()
```
